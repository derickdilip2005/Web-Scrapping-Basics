```python
import pandas
import requests
from bs4 import BeautifulSoup
```

```python
r = requests.get('https://books.toscrape.com/')
print(r)
print(r.content)
```

```python
books = BeautifulSoup(r.text, 'html.parser')
books
```

```python
for book in books.find_all('a'):
    bookt = book.get('title')
    if bookt is not None:
        print(bookt)
```

```python
for price in books.find_all('p', class_='price_color'):
    print(price.text)
```

```python
for availability in books.find_all('p', class_='instock availability'):
    print(availability.text)
```

```python
books = []
price = []
availability = []

for i in range(1,51):
  r = requests.get(f"https://books.toscrape.com/catalogue/page-{i}.html")
  book = BeautifulSoup(r.text, 'html.parser')

  for i in book.find_all('a'):
    bookt = i.get('title')
    if bookt is not None:
      books.append(bookt)

  for j in book.find_all('p', class_='price_color'):
    price.append(j.text)

  for k in book.find_all('p', class_='instock availability'):
    availability.append(k.text)
```

```python
for i in books:
  print(i)
print(len(books))
```

```python
for i in price:
  print(i)
print(len(price))
```

```python
for i in availability:
  print(i)
print(len(availability))
```

```python
df = pandas.DataFrame({'Book': books, 'Price': price, 'Availability': availability})
df
```
